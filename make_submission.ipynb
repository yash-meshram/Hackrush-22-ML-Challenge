```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import joblib
import math
```

```python
model = joblib.load("model_cat_4.joblib")
```

```python
df = pd.read_csv('test.csv')
```

```python
type(df['Timestep'])
```

```python
df['Year'] = df['Timestep'].apply(lambda x: int(x[2:6]))
```

```python
Course_dummies = pd.get_dummies(df['Course'])
Course_dummies
```

```python
df = pd.concat([df, Course_dummies.drop('ZZ23306', axis = 'columns')], axis='columns')
df
```

```python
Faculty_dummies = pd.get_dummies(df['Faculty'])
Faculty_dummies
```

```python
df = pd.concat([df, Faculty_dummies.drop('Prof. Zoie Mueller', axis = 'columns')], axis='columns')
df
```

```python
min_year = min(df['Year'])
max_year = max(df['Year'])

df['Year_'] = df['Year'].apply(lambda x: ( x - min_year ) / ( max_year - min_year ))
```

```python
df.dtypes
```

```python
x = df.drop(['Id', 'Timestep', 'Course', 'Faculty', 'Year_'], axis = 'columns')
```

```python
y_pred = model.predict(x)
```

```python
df['Predicted'] = y_pred+25
```

```python
df_submission = df[['Id', 'Predicted']]
df_submission
```

```python
df_submission.to_csv("submission_best.csv", index=False)
```
